```python
import numpy as np
import pandas as pd
import yfinance as yf
import plotly.graph_objs as go

def data(coin_name, period, interval):
    data = yf.download(tickers=coin_name, period = period, interval = interval)
    fig = go.Figure()
    fig.add_trace(go.Candlestick(x=data.index,
                    open=data['Open'],
                    high=data['High'],
                    low=data['Low'],
                    close=data['Close'], name = 'market data'))
    fig.update_layout(
        title='Live coin share price evolution',
        yaxis_title='Price per dollar')
    fig.update_xaxes(
        rangeslider_visible=True,
        rangeselector=dict(
            buttons=list([
                dict(count=15, label="15m", step="minute", stepmode="backward"),
                dict(count=45, label="45m", step="minute", stepmode="backward"),
                dict(count=1, label="HTD", step="hour", stepmode="todate"),
                dict(count=6, label="6h", step="hour", stepmode="backward"),
                dict(step="all")
            ])
        )
    )
    return fig.show()
```

```python
data('BTC-USD','240h','15m')
```
